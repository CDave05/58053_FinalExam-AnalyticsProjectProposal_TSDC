```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files 
uploaded = files.upload()
```

```python
import io
```

```python
df = pd.read_csv(io.StringIO(uploaded['MetroManilaCases.csv'].decode('utf-8')))
df
```

```python
sns.barplot(x = 'Cumulative Total Confirmed Cases',y = 'City', data = df, ci = None, order=df.sort_values('Cumulative Total Confirmed Cases',ascending = False).City);

plt.xlabel('Cumulative Total Confirmed Cases', fontsize=22)
plt.ylabel('City', fontsize=22)
plt.tick_params(labelsize=10)
```

```python
sns.barplot(x = 'Cumulative Total Recovered Cases',y = 'City', data = df, ci = None, order=df.sort_values('Cumulative Total Recovered Cases',ascending = False).City);

plt.xlabel('Cumulative Total Recovered Cases', fontsize=22)
plt.ylabel('City', fontsize=22)
plt.tick_params(labelsize=10)
```

```python
sns.barplot(x = 'Cumulative Total Recovered Cases',y = 'City', data = df, ci = None, order=df.sort_values('Cumulative Total Recovered Cases',ascending = False).City);

plt.xlabel('Cumulative Total Recovered Cases', fontsize=22)
plt.ylabel('City', fontsize=22)
plt.tick_params(labelsize=10)
```

```python
sns.barplot(x = 'Cumulative Total Expired Cases',y = 'City', data = df, ci = None, order=df.sort_values('Cumulative Total Expired Cases',ascending = False).City);

plt.xlabel('Cumulative Total Expired Cases', fontsize=22)
plt.ylabel('City', fontsize=22)
plt.tick_params(labelsize=10)
```

```python
sns.barplot(x = 'Current Active Cases',y = 'City', data = df, ci = None, order=df.sort_values('Current Active Cases',ascending = False).City);

plt.xlabel('Current Active Cases', fontsize=22)
plt.ylabel('City', fontsize=22)
plt.tick_params(labelsize=10)

```
